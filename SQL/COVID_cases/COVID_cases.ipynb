-- Общее количество случаев в сравнении с общим количеством смертей.

-- Вероятность смерти, если вы заразитесь covid в России.

```python
sql = '''SELECT location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 as probability_of_death
FROM COVID_deaths
WHERE location like '%Russia%' and continent is not null
ORDER BY 2
'''
```

```python
select(sql)
```

-- Общее количество случаев в сравнении с населением 
-- Какой процент населения инфицирован Covid

```python
sql = ''' SELECT location, date, total_cases, population, (total_cases/population)*100 AS infected_persentage
FROM COVID_deaths
WHERE location like '%%russia%' and continent is not null
ORDER BY 1,2 DESC
LIMIT 5
'''
```

```python
select(sql)
```

-- Страны с самым высоким уровнем инфицирования по сравнению с населением

```python
sql = ''' SELECT location, MAX(total_cases), population, MAX((total_cases/population)*100) AS population_infected
FROM COVID_deaths
WHERE continent is not null
GROUP BY location, population
ORDER BY population_infected DESC
LIMIT 10;
'''
```

```python
select(sql)
```

-- Страны с наибольшим количеством смертей на душу населения

```python
sql = ''' SELECT location, max(cast(total_deaths as int)) AS death_per_country, MAX((total_deaths/population)*100) AS death_rate
FROM COVID_deaths
WHERE continent is not null
GROUP BY location
ORDER BY death_rate desc
LIMIT 10
'''
```

```python
select(sql)
```

-- Континент с наибольшим количеством смертей на душу населения

```python
sql = ''' SELECT continent, max(cast(total_deaths as int)) AS death_per_continent
FROM COVID_deaths 
WHERE continent is not null
GROUP BY continent 
ORDER BY death_per_continent DESC;
'''
```

```python
select(sql)
```

-- Общая численность населения по сравнению с прививитыми.

--(найдем общее кол-во вакцинированных в каждой стране)

```python
sql = '''
SELECT dea.continent, dea.location, dea.date, dea.population, vacc.new_vaccinations,
SUM(cast(vacc.new_vaccinations as int)) OVER (Partition by dea.location ORDER BY dea.location, dea.date) as Vaccinated_per_country
FROM COVID_deaths dea join COVID_vacc vacc 
    on dea.location = vacc.location 
    and dea.date = vacc.date
WHERE dea.continent is not null
'''
```

```python
select(sql)
```

```python
-- Общая численность населения по сравнению с прививитыми.(CTE)

--(найдем общее кол-во вакцинированных в каждой стране)
```

```python
sql = '''With popVSvac (continent, location, date, population, new_vaccinations,Vaccinated_per_country) 
as (
SELECT dea.continent, dea.location, dea.date, dea.population, vacc.new_vaccinations,
SUM(cast(vacc.new_vaccinations as int)) OVER (Partition by dea.location ORDER BY dea.location, dea.date) as Vaccinated_per_country
FROM COVID_deaths dea join COVID_vacc vacc 
    on dea.location = vacc.location 
    and dea.date = vacc.date
WHERE dea.continent is not null
) 
SELECT *, (Vaccinated_per_country/population)*100 as percentage_of_vaccineted 
FROM popVSvac
order by 5 desc'''
```

```python
select(sql)
```
